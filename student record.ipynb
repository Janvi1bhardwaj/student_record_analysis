```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
a=pd.read_csv("student record.csv")
a.head()
```

```python
a.describe()
```

```python
a.info()
```

```python
a.isnull().sum()
```

```python
a=a.drop("Unnamed: 0",axis=1)
print(a.head())
```

```python
a["WklyStudyHours"]=a["WklyStudyHours"].replace("10-May","5-10")
a
```

# gender distributon

```python
plt.figure(figsize=(5,5))
s=sns.countplot(x="Gender",data=a,palette="Accent")
for container in s.containers:
    s.bar_label(container)
plt.title("Gender Distribution")
plt.savefig("Gender Distribution.jpg")
plt.show()
```

```python
g=a.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(g)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(g,annot=True)
plt.title("ParentEduc Impact")
plt.savefig("ParentEduc Impact.jpg")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(m,annot=True)
plt.title("Parent Marital Status Impact")
plt.savefig("Parent Marital Status Impact.jpg")
plt.show()
```

```python
print(a["EthnicGroup"].unique())
```

```python
groupA_count = a.loc[a['EthnicGroup'] == "group A"].shape[0]
groupB_count = a.loc[a['EthnicGroup'] == "group B"].shape[0]
groupC_count = a.loc[a['EthnicGroup'] == "group C"].shape[0]
groupD_count = a.loc[a['EthnicGroup'] == "group D"].shape[0]
groupE_count = a.loc[a['EthnicGroup'] == "group E"].shape[0]

mlist = [groupA_count, groupB_count, groupC_count, groupD_count, groupE_count]
labels = ["group A", "group B", "group C", "group D", "group E"]
plt.figure(figsize=(5,4))
plt.pie(mlist, labels=labels, autopct="%.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.savefig("Distribution of Ethnic Groups.jpg")
plt.show()
```

```python
plt.figure(figsize=(7,5))
s=sns.countplot(x="EthnicGroup",data=a,palette="Accent")
for container in s.containers:
    s.bar_label(container)
plt.title("Count of Each Ethnic Group")
plt.savefig("Count of Each Ethnic Group.jpg")
plt.show()
```
